**Question 12:** Write a Python program to demonstrate comparison operators: equal to, not equal to, greater than, less than.

```python
a = int(input("Enter the number :"))
b = int(input("Enter the Second number :"))

if( a == b):
    print("qual to")
elif( a != b):
    print(" not equal to")
elif( a > b):
    print("greater than")
elif( a < b ):
    print("less than")
else:
    print("invalid input")

```
